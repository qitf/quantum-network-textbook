# 光子が伝送路中でロスする場合

- 物理現象としてはどういう過程かは置いといて、10%の確率でロスする、90%しかE2Eで光子が直接到達しない、というのを統計的に示せば良いだけ
- 量子回路の外側の話
- 逆にエンタングルメントスワッピングした時に、エンタングルメントスワッピングがない時と何を比べたいかを明らかにしておく必要がある
- 直接伝送した時とエンタングルメントスワッピングした時とでもつれを共有した時とで、共有の成功確率(もしくは失敗確率)を距離の関数で描けば良いのではないか
    - 光ファイバーの伝送損失をもとに距離に依存したドロップ確率を算出
    - なんかそれはシミュレートせずとも解析的に求まってしまうな
    - エンタングルメントスワッピングもベル測定やへラルディングが必ず成功すると仮定したら、結局ロスの要因はそこに到達するまでのファイバーの損失になる
    - 必ず指定した距離の中央に中継器を置くのだとしたら、直接伝送と比較して同じ距離でも半分の距離の確率でエンタングルメント共有ができることになる
    - だから、統計を出すという点だと面白みに欠ける
- ただ、実際にネットワークシミュレータという観点で、「あ、つながった」「あ、届かなかった」というのは面白いと思う。
    - 直接伝送だと、体感として距離を伸ばすとコネクション張ろうと思ってもドロップすることが多いなって感じられるかもしれない
    - 一方で中継器があると比較的コネクション張れるな、とか

## シミュレータ作る時の懸念点

- 光子が動く様子とか、途中でなくなる様子とか、もしくは流石にくるだろって時間待っても来ない(タイムアウトする)場合
    - 損失確率の時間発展や距離発展を計算する必要がある
- 測定した結果がどの状態か
    - 状態へのノイズや状態の時間発展をシミュレート
    - どちらかがドロップしたらもう測定しないということにする
    - どちら生き残った時に初めて測定する
